```python
from mip.constants import OptimizationStatus
from scipy.spatial import distance_matrix
from sklearn.datasets import make_moons

from location_allocation import (
    MaximizeCoverage,
    MaximizeCoverageCapacitated,
    MaximizeCoverageMinimizeCost,
    MaximizeCoverageMinimizeFacilities,
)
import utils

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
plt.rcParams.update({"figure.figsize": [12, 8] })


def make_colors(status):
    colors = []
    for s in status:
        if s == OptimizationStatus.OPTIMAL:
            colors.append("green")
        elif s == OptimizationStatus.FEASIBLE:
            colors.append("orange")
        elif s == OptimizationStatus.INFEASIBLE:
            colors.append("salmon")
        elif s == OptimizationStatus.NO_SOLUTION_FOUND:
            colors.append("red")
    return colors
```

## Problem Setup

```python
n_points = [100, 300, 500, 700, 1000, 1500]
custom_markers = [
    Line2D(range(1), range(1), color="white", markerfacecolor="green", marker="o", markersize=7),
    Line2D(range(1), range(1), color="white", markerfacecolor="orange", marker="o", markersize=7),
    Line2D(range(1), range(1), color="white", markerfacecolor="salmon", marker="o", markersize=7),
    Line2D(range(1), range(1), color="white", markerfacecolor="red", marker="o", markersize=7),
]

max_mip_gap = 0.1
max_seconds = 2000

problems = [
    {
        "title": "Maximize Coverage (points, facilities, facilities to be sited)",
        "filename": "maximize_coverage.png",
    },
    {
        "title": "Maximize Capacitated Coverage (points, facilities, facilities to be sited)",
        "filename": "maximize_coverage_capacitated.png",
    },
    {
        "title": "Maximize Coverage Minimize Facilities (points, facilities, facilities to be sited)",
        "filename": "maximize_coverage_minimize_facilities.png",
    },
    {
        "title": "Maximize Coverage Minimize Cost (points, facilities, facilities to be sited)",
        "filename": "maximize_coverage_minimize_cost.png",
    },
]
```

## Run Optimization

```python
for i, problem in enumerate(problems):

    x = []
    y = []
    status = []
    for row_idx, n in enumerate(n_points):

        points, _ = make_moons(n, noise=0.15)
        facilities = utils.generate_candidate_facilities(points, n / 50)
        dist_matrix = distance_matrix(points, facilities)
        facilities_to_site = round(len(facilities) / 2)
        if i == 0:
            p = MaximizeCoverage(
                points,
                facilities,
                dist_matrix,
                dist_cutoff=0.2,
                facilities_to_site=facilities_to_site,
                max_gap=max_mip_gap,
            ).optimize(max_seconds=max_seconds)
        elif i == 1:
            p = MaximizeCoverageCapacitated(
                points,
                facilities,
                dist_matrix,
                capacities=utils.generate_facility_capacities(len(facilities)),
                dist_cutoff=0.2,
                facilities_to_site=facilities_to_site,
                max_gap=max_mip_gap,
            ).optimize(max_seconds=max_seconds)
        if i == 2:
            p = MaximizeCoverageMinimizeCost(
                points,
                facilities,
                dist_matrix,
                dist_cutoff=0.2,
                facilities_to_site=facilities_to_site,
                max_gap=max_mip_gap,
            ).optimize(max_seconds=max_seconds)
        elif i == 3:
            p = MaximizeCoverageMinimizeFacilities(
                points,
                facilities,
                dist_matrix,
                dist_cutoff=0.2,
                max_gap=max_mip_gap,
            ).optimize(max_seconds=max_seconds)
        print(p.model.status)

        y.append(float("{:.2f}".format(p.result.time_elapsed)))
        status.append(p.model.status)

    colors = make_colors(status)
    x = list(range(0, len(colors)))
    ax = plt.subplot()
    ax.cla()
    ax.title.set_text(problem["title"])
    ax.scatter(x=x, y=y, s=40, c=colors)
    plt.xticks(x, ["A", "B", "C", "D", "E", "F"])
    ax.set_xlabel("Experiments (maximum MIP gap = 0.1)")
    ax.set_ylabel("Elapsed time (in seconds, max = 2000s)")
    ax.set_xmargin(0.2)
    for i, x_val in enumerate(x):
        plt.annotate(
            f"{n_points[i]} / {int(n_points[i]/50)} / {int(n_points[i]/50/2)}",
            (i + 0.1, y[i]),
            fontsize=10,
            bbox=dict(boxstyle="round", fc=(1.0, 1.0, 1.0), ec="black"),
        )
    ax.legend(
        custom_markers,
        ["Optimal", "Feasible", "Infeasible", "No solution found"],
        loc="upper left",
        facecolor="white",
        framealpha=1,
    )
    plt.show()
    #plt.savefig(problem["filename"], dpi=300)
```
